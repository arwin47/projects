### Recommendation of Persomnalization Themes based on user based Collaborative Filtering

The [dataset](https://www.kaggle.com/datasets/alfarisbachmid/personalized-recommendation-systems-dataset/data) available in kaggle has User ID and Item ID and the respective ratings. Based on similarity of a user to other users, a user will get recommendations by looking at the similarity and the best rated interests. This recommendation systen does consider the existing interests of the user and includes those categories too in the recommendations.

#### Creating dataframe with a row for each user

```python
matrix = pd.pivot_table(data,index='User_ID', columns='Item_ID',values='Rating').fillna(0)
matrix.iloc[:5,:5]
```

There ae 5000 users and 2000 items

```python
matrix.shape
```

#### Creating Cosine Similarity Scores

```python
similarity = cosine_similarity(matrix)
similarity.shape
```

```python
similarity_df = pd.DataFrame(similarity,index=matrix.index,columns=matrix.index)
similarity_df.iloc[:5,:5]
```

#### Creating a Recommendation logic
For a user,
1. Using the Cosine similarity, most similar users are selected.
2. Users with similarity > 0 are selected.
3. Items best rated by each user is selected.
4. Best rated item categories are selected for recommendations.

```python
def recommend(user,similarity = similarity_df,pivot = matrix, n_recoms = 3):
    similar_users = similarity_df[user].drop(user).sort_values(ascending=False)
    similar_users = similar_users[similar_users > 0]
    recommendations = {}

    for user, score in similar_users[:5].items():
        rated_items = matrix.loc[user].sort_values(ascending=False)
        rated_items = rated_items[rated_items >0].items()
        for interest,rating in rated_items:
            recommendations[interest] = recommendations.get(interest,0) + (score*rating)

    recommendations = sorted(recommendations.items(),key=lambda items: items[1],reverse=True)

    all_recommendations = []
    for item, score in recommendations:
        all_recommendations.append(item)
    uniques = []
    for category in list(data[(data['Item_ID'].isin(all_recommendations))]['Category']):
        if category not in uniques:
            uniques.append(category)    
        
    return uniques[:n_recoms]
```

#### Generating Recommendations

```python
recommend("User_1011")
```
