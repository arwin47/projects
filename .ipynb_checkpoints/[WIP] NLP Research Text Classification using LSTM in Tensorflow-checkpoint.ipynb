```python
def prep_df_from_txt(filename, directory):
    with open(directory+filename, "r") as file:
        lines = file.readlines()  
        master_list = []
        inner_list = []
        for line in lines:
            if line.strip() != "":
                inner_list.append(line)
            else:
                master_list.append(inner_list)
                inner_list = []
    
    samples = []       
    for inner_list in master_list:
        for num,line in enumerate(inner_list[1:]):
            collect = {}
            splits = line.split("\t")
            target = splits[0]
            text = splits[1].strip().lower()
            collect['target'] = target
            collect['text'] = text
            collect['line_number'] = num + 1 
            collect['total_lines'] = len(inner_list)-1
            samples.append(collect)
    sample_df = pd.DataFrame(samples)
    return sample_df
```

```python
# Token and char embeddings combined model

token_inputs = layers.Input(shape=(),dtype=tf.string,name='token_input')
token_vectors = text_vectorizer(token_inputs)
token_embeddings = vector_embed(token_vectors)
token_conv = layers.Conv1D(64, kernel_size=5,padding='same',activation='relu')(token_embeddings)
token_avg_pool = layers.GlobalAveragePooling1D()(token_conv)
token_output = layers.Dense(128,activation='relu')(token_avg_pool)
token_model=tf.keras.Model(inputs=token_inputs,outputs=token_output)


char_inputs = layers.Input(shape=(),dtype=tf.string,name='char_input')
char_vectors = char_vectorizer(char_inputs)
char_embeddings = char_embedding(char_vectors)
char_lstm = layers.Bidirectional(layers.LSTM(25))(char_embeddings)
char_output = layers.Dense(128,activation='relu')(char_lstm)
char_model=tf.keras.Model(inputs=char_inputs,outputs=char_output)

#Concatenation
token_char_concat = layers.Concatenate(name='combined_input')([token_model.output, char_model.output])


first_drop_out = layers.Dropout(0.5)(token_char_concat)
dense = layers.Dense(200,activation="relu")(first_drop_out)
second_drop_out = layers.Dropout(0.5)(dense)
outputs = layers.Dense(int(len(label_encoder.classes_)),activation="softmax")(second_drop_out)


model_3 = tf.keras.Model(inputs=[token_model.input,char_model.input],outputs=outputs, name="combined_word_and_char_tokenized")

model_3.compile(loss="categorical_crossentropy", optimizer= tf.keras.optimizers.Adam(),metrics=['accuracy'])

model_3.summary()
```

```python
train_token_char_data = tf.data.Dataset.from_tensor_slices((train_sentences,train_chars))
val_token_char_data = tf.data.Dataset.from_tensor_slices((val_sentences,val_chars))

train_token_char_labels = tf.data.Dataset.from_tensor_slices(train_labels_one_hot)
val_token_char_labels = tf.data.Dataset.from_tensor_slices(val_labels_one_hot)

train_token_char_dataset = tf.data.Dataset.zip((train_token_char_data,train_token_char_labels)).batch(32).prefetch(tf.data.AUTOTUNE)
val_token_char_dataset = tf.data.Dataset.zip((val_token_char_data,val_token_char_labels)).batch(32).prefetch(tf.data.AUTOTUNE)

```

```python
# Token and char embeddings + Positional (Line Numbers) combined model

token_inputs = layers.Input(shape=(),dtype=tf.string,name='token_input')
token_vectors = text_vectorizer(token_inputs)
token_embeddings = vector_embed(token_vectors)
token_conv = layers.Conv1D(64, kernel_size=5,padding='same',activation='relu')(token_embeddings)
token_avg_pool = layers.GlobalAveragePooling1D()(token_conv)
token_output = layers.Dense(128,activation='relu')(token_avg_pool)
token_model=tf.keras.Model(inputs=token_inputs,outputs=token_output)


char_inputs = layers.Input(shape=(),dtype=tf.string,name='char_input')
char_vectors = char_vectorizer(char_inputs)
char_embeddings = char_embedding(char_vectors)
char_lstm = layers.Bidirectional(layers.LSTM(25))(char_embeddings)
char_output = layers.Dense(128,activation='relu')(char_lstm)
char_model=tf.keras.Model(inputs=char_inputs,outputs=char_output)

line_number_inputs = layers.Input(shape=(15,),dtype=tf.int32, name="line_number_input")
line_number_outputs = layers.Dense(32,activation = "relu")(line_number_inputs)
line_number_model = tf.keras.Model(inputs=line_number_inputs,outputs=line_number_outputs)

total_lines_inputs = layers.Input(shape=(20,),dtype=tf.int32, name="total_lines_input")
total_lines_outputs = layers.Dense(32,activation = "relu")(total_lines_inputs)
total_lines_model = tf.keras.Model(inputs=total_lines_inputs,outputs=total_lines_outputs)


token_char_combined = layers.Concatenate()([token_model.output,char_model.output])

combined_dense = layers.Dense(256, activation="relu")(token_char_combined)
combined_drop_out = layers.Dropout(0.5)(combined_dense)

# Combining the concatenated Token and Char Embedded models with positional models

token_char_position_combined = layers.Concatenate()([line_number_model.output,total_lines_model.output,combined_drop_out])

output_layer = layers.Dense(5,activation="softmax")(token_char_position_combined)


model_4 = tf.keras.Model(inputs=[line_number_model.input,total_lines_model.input,
                                token_model.input,char_model.input], outputs=output_layer)

model_4.compile(loss='categorical_crossentropy',optimizer=tf.keras.optimizers.Adam(),metrics=['accuracy'])
model_4.summary()
```

```python


# Create training and validation datasets (all four kinds of inputs)
train_pos_char_token_data = tf.data.Dataset.from_tensor_slices((train_line_numbers_one_hot, # line numbers
                                                                train_total_lines_one_hot, # total lines
                                                                train_sentences, # train tokens
                                                                train_chars)) # train chars
train_pos_char_token_labels = tf.data.Dataset.from_tensor_slices(train_labels_one_hot) # train labels
train_pos_char_token_dataset = tf.data.Dataset.zip((train_pos_char_token_data, train_pos_char_token_labels)) # combine data and labels
train_pos_char_token_dataset = train_pos_char_token_dataset.batch(32).prefetch(tf.data.AUTOTUNE) # turn into batches and prefetch appropriately

# Validation dataset
val_pos_char_token_data = tf.data.Dataset.from_tensor_slices((val_line_numbers_one_hot,
                                                              val_total_lines_one_hot,
                                                              val_sentences,
                                                              val_chars))
val_pos_char_token_labels = tf.data.Dataset.from_tensor_slices(val_labels_one_hot)
val_pos_char_token_dataset = tf.data.Dataset.zip((val_pos_char_token_data, val_pos_char_token_labels))
val_pos_char_token_dataset = val_pos_char_token_dataset.batch(32).prefetch(tf.data.AUTOTUNE) # turn into batches and prefetch appropriately

# Check input shapes
train_pos_char_token_dataset, val_pos_char_token_dataset
```
