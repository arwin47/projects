```python
from langchain_community.agent_toolkits.load_tools import load_tools
from langchain.agents import create_agent
from langchain_ollama import ChatOllama
import pprint

from langchain_community.tools import DuckDuckGoSearchResults
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
# from langchain.agents.executor import AgentExecutor
```

```python
search = DuckDuckGoSearchResults()
wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())

tools = [search,wikipedia]

llm = ChatOllama(model="mistral",temperature=0)
tools = load_tools(["llm-math","wikipedia","arxiv"],llm=llm)
tools.append(search)

agent = create_agent(tools= tools, model=llm, system_prompt="you are a helpful assistant with quick math skills")

# agent_executor = AgentExecutor(
#     agent=agent,
#     tools=tools,
#     verbose=True,                
#     return_intermediate_steps=True 
# )
user_question = input("How can I help?")

try:
    response = agent.invoke( {"messages": [{"role": "user", "content": user_question}]} )
    pprint.pprint(response)
except Exception as e:
    print(f"An error occured: {e}")
```

```python
def print_reasoning_chain(response):
    steps = response.get("intermediate_steps", [])
    if not steps:
        print("No intermediate steps found. Set return_intermediate_steps=True on AgentExecutor.")
        return

    for i, (action, observation) in enumerate(steps, start=1):
        print(f"Step {i}:")
        print(f"  Tool: {action.tool}")
        print(f"  Tool input: {action.tool_input}")
        if getattr(action, "log", None):
            print("  Agent log:")
            print("  " + str(action.log).replace("\n", "\n  "))
        print(f"  Observation: {observation}")
        print("-" * 60)

print_reasoning_chain(response)
```
