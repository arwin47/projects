### AI Agent using Langchain Tools

This demonstrates the usage of common Langchain tools if the LLM is not capable of producing the required result

#### Importing the required libraries

```python
from langchain_community.tools import DuckDuckGoSearchResults
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain.agents import create_agent
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama import ChatOllama
```

```python
llm = ChatOllama(model="llama3.1:8b",
                temperature=0)
```

```python
search = DuckDuckGoSearchResults()
wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())


tools = []
```

```python
agent = create_agent(llm,tools,system_prompt="You are a helpful research assistant.")

result = agent.invoke({    "messages": [
        {"role": "user", "content": "india south aftica test series 2025"}
    ]})
```

```python
result
```

#### Results without tools

The response reads 'The series is scheduled for 2025, which means it's still a year or two away' which means the LLM alone doesn't have the recent updates on the topic. It would need the help of tools to check online and get the recent results.

#### Adding Tools

```python
tools = [search,wikipedia]
```

```python
agent = create_agent(llm,tools,system_prompt="You are a helpful research assistant.")

result = agent.invoke({    "messages": [
        {"role": "user", "content": "who won the first match in india south aftica test series 2025"}
    ]})
result
```

#### Results with tools

The response reads 'I was unable to find information on who won the first match in the India vs South Africa test series 2025. The provided Wikipedia pages do not contain any information about the specific matches played during this tour' which means the LLM was able to check Wikipedia but not able to produce the required result. Let us see if we can force it to use search through prompts.

```python
agent = create_agent(llm,tools,system_prompt="You are a helpful research assistant. Use search tool provided to answer the question.")

result = agent.invoke({    "messages": [
        {"role": "user", "content": "who won the first match in india south aftica test series 2025"}
    ]})
result
```
